```python
from keras.models import load_model

classifier = load_model('emojify.h5')
```

```python
import cv2
import numpy as np
import os
import IPython.display as ipd
import webbrowser as wb

emojify = {"[0]": "Angry", 
           "[1]": "Happy",
           "[2]": "Neutral",
           "[3]": "Sad",
           "[4]": "Shock"}

face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def face_detector(img):

    # Convert image to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is ():
        return img, []

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        cropped_image = img[y:y+h, x:x+w]
        cropped_image = cv2.resize(cropped_image, (224, 224))
    return img, cropped_image


# Open Webcam
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    image, face = face_detector(frame)

    try:
        input_im = face
        input_original = input_im.copy()
        input_original = cv2.resize(input_original, None, fx=0.5, fy=0.5, interpolation = cv2.INTER_LINEAR)

        input_im = cv2.resize(input_im, (224, 224), interpolation = cv2.INTER_LINEAR)
        input_im = input_im / 255.
        input_im = input_im.reshape(1,224,224,3) 
        # res contain 
        res = np.argmax(classifier.predict(input_im, 1, verbose = 0), axis=1)
        if res[0] == 0 :
            img2 = cv2.imread('C:/Users/sdeo7/OneDrive/Desktop/Folders/Project_MachineLearning/Emojis/angry.jpg')
            rows,cols,channels = img2.shape
            roi = image[0:rows, 0:cols] # To put the emoji on top-left corner
            img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
            ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY) # Creating Mask of emoji
            mask_inv = cv2.bitwise_not(mask)
            img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv) # Blacking out the area of emoji in the roi
            img2_fg = cv2.bitwise_and(img2,img2,mask = mask) # Take only region of emoji from emoji image
            dst = cv2.add(img1_bg,img2_fg)
            image[0:rows, 0:cols ] = dst # Modifying main image
            emotion = emojify[str(res)]
            cv2.putText(image, emotion, (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Emojify', image)            
        elif res[0] == 1 :
            img2 = cv2.imread('C:/Users/sdeo7/OneDrive/Desktop/Folders/Project_MachineLearning/Emojis/happy.jpg')
            rows,cols,channels = img2.shape
            roi = image[0:rows, 0:cols] # To put the emoji on top-left corner
            img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
            ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY) # Creating Mask of emoji
            mask_inv = cv2.bitwise_not(mask)
            img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv) # Blacking out the area of emoji in the roi
            img2_fg = cv2.bitwise_and(img2,img2,mask = mask) # Take only region of emoji from emoji image
            dst = cv2.add(img1_bg,img2_fg)
            image[0:rows, 0:cols ] = dst # Modifying main image
            emotion = emojify[str(res)]
            cv2.putText(image, emotion, (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Emojify', image)  
        elif res[0] == 2 :
            img2 = cv2.imread('C:/Users/sdeo7/OneDrive/Desktop/Folders/Project_MachineLearning/Emojis/neutral.png')
            rows,cols,channels = img2.shape
            roi = image[0:rows, 0:cols] # To put the emoji on top-left corner
            img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
            ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY) # Creating Mask of emoji
            mask_inv = cv2.bitwise_not(mask)
            img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv) # Blacking out the area of emoji in the roi
            img2_fg = cv2.bitwise_and(img2,img2,mask = mask) # Take only region of emoji from emoji image
            dst = cv2.add(img1_bg,img2_fg)
            image[0:rows, 0:cols ] = dst # Modifying main image
            emotion = emojify[str(res)]
            cv2.putText(image, emotion, (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Emojify', image)   
        elif res[0] == 3 :
            img2 = cv2.imread('C:/Users/sdeo7/OneDrive/Desktop/Folders/Project_MachineLearning/Emojis/sad.png')
            rows,cols,channels = img2.shape
            roi = image[0:rows, 0:cols] # To put the emoji on top-left corner
            img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
            ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY) # Creating Mask of emoji
            mask_inv = cv2.bitwise_not(mask)
            img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv) # Blacking out the area of emoji in the roi
            img2_fg = cv2.bitwise_and(img2,img2,mask = mask) # Take only region of emoji from emoji image
            dst = cv2.add(img1_bg,img2_fg)
            image[0:rows, 0:cols ] = dst # Modifying main image
            emotion = emojify[str(res)]
            cv2.putText(image, emotion, (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Emojify', image) 
        elif res[0] == 4 :
            img2 = cv2.imread('C:/Users/sdeo7/OneDrive/Desktop/Folders/Project_MachineLearning/Emojis/shock.png')
            rows,cols,channels = img2.shape
            roi = image[0:rows, 0:cols] # To put the emoji on top-left corner
            img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
            ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY) # Creating Mask of emoji
            mask_inv = cv2.bitwise_not(mask)
            img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv) # Blacking out the area of emoji in the roi
            img2_fg = cv2.bitwise_and(img2,img2,mask = mask) # Take only region of emoji from emoji image
            dst = cv2.add(img1_bg,img2_fg)
            image[0:rows, 0:cols ] = dst # Modifying main image
            emotion = emojify[str(res)]
            cv2.putText(image, emotion, (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Emojify', image) 
        
        print(res)

    except:
        print("!!Face Not Found!!")
        pass

    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break

cap.release()
cv2.destroyAllWindows()
```
